# **Сравнение подходов к построению системы рекомендаций**

Требуется сравнить два подхода к построению рекомендаций: коллаборативный и коллаборативный + контентный.



1.   **Выбор метрики**

Для начала необходимо выбрать метрику, по которой будем сравнивать рекомендации. Можно придумать множество критериев качества рекомендательной системы. Например, разнообразие списка рекомендаций, новизна предложенных объектов, неожиданность и неочевидность рекомендации. Я предлагаю остановиться на метрике, которая будет в первую очередь измерять полезность пользователя. 

Чаще всего для того, чтобы оценить полезность рекоммендации для пользователя, считают ошибку предсказанной системой оценки и реальной оценки пользователя. Например, RMSE, MSE, также можно отнести такие метрики, как Precision и Recall. 

Минусом данного подхода является то, что они не учитывают относительные позиции объектов в списке рекоммендаций, в то время как это является важным критерием для пользователя. Пользователь в первую очередь смотрит на верхнюю часть списка рекоммендаций и хочет сразу найти подходящий для него объект, не листая бесконечно долго вниз. 

Наиболее популярными метриками, учитывающими ранжирование, являются коэффициенты ранговой корреляции Спирмена и Кендалла. Я выбрала коэффициент корреляции Кендалла (тау Кендалла), так как он более устойчивый для большого количества данных и вычисляется быстрее, чем коэффициент Спирмена.

Так как у нас есть данные о времени выставления оценки пользователями, мы можем использовать их для построения тестовой выборки. 
В реальном мире рекомендательная система получает данные на вход в онлайн режиме, поэтому должна хорошо справляться с данными, поступающими по времени. Отсортируем данные по времени и уберем 10% последних наблюдений.


```python
df_sorted = df.sort_values(by='timestamp')
```

```python
train, test = np.split(df_sorted, [int(0.9 * len(df))])
train.shape
```

```python
X_train = train.drop(['rating', 'timestamp'], axis=1)
y_train = train[['rating']]
X_test =  test.drop(['rating', 'timestamp'], axis=1)
y_test = test[['rating']]
```
